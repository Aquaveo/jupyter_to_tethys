```python
import param
import parambokeh
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

output_notebook()
```

```python
from bokeh.models import ColumnDataSource       
from bokeh.plotting import figure

# This is my exmpale shows how to use standard bokeh plot with parambokeh widgets
class BokehSine(param.Parameterized):
    
    # define param parameters
    magnitude = param.Number(1, bounds=(-5, 5))
    offset = param.Number(default=0.0, bounds=(-5.0,5.0))    
    phase = param.Number(default=0.0,bounds=(0.0,2*np.pi))
    frequency = param.Number(default=1.0, bounds=(0.1, 5.1))
    N = param.Integer(default=200, bounds=(0,None))
    
    # placeholder for parambohek plot!!!!!!
    output = parambokeh.view.Plot()
    
    def update(self, **kwargs): 
                      
        # construct a Bokeh Plot
        x = np.linspace(0, 4*np.pi, self.N)
        y = self.magnitude * np.sin(self.frequency * x + self.phase) + self.offset
        source = ColumnDataSource(data=dict(x= x, y=y))
       
        # Or do this instead
        #x,y = self.XY()
        #source = ColumnDataSource(data=dict(x=x, y=y))        
        
        plot = figure(plot_height=400, plot_width=400,
              tools="crosshair,pan,reset,save,wheel_zoom",
              x_range=[0, 4*np.pi], y_range=[-2.5, 2.5])
        plot.line('x', 'y', source=source, line_width=3, line_alpha=0.6)  
        
        # Put it into parambohek plot placeholder
        self.output = plot
     
    def XY(self,**params):
        p = param.ParamOverrides(self,params)
        x = np.linspace(0, 4*np.pi, p.N)
        y = p.amplitude*np.sin(p.frequency*x)
        return x,y   

example = BokehSine(name='BokehSine')
layout = parambokeh.Widgets(example, on_init=True, callback=example.update, view_position='right')
```
