Now embed a bokeh server and start a bokeh app

```python
from bokeh.io import show, output_notebook
output_notebook()
```

```python
from bokeh.layouts import widgetbox
from bokeh.models.widgets import Slider
import numpy
import time

# This is a bokeh app! Should run with a bokeh server!!
# This bokeh can run outside of jupyter as well!
# bokeh widget on_change event (bokeh widget python callback) only works with a bokeh server
def modify_doc(doc):    
    
    my_slider_bokeh = Slider(start=0, end=100, value=1, step=.1, title="Stuff")
    
    # Python callback to response to UI changes
    def UI_changed(attr, old, new):        
        print ("{}:{}->{}".format(attr, old, new))  # This outputs to widget UI via bokeh websocket
        print ("python obj value: {0}".format(my_slider_bokeh.value))         
    
    # A func to demo changing python Obj
    def update_python_obj():                   
        my_slider_bokeh.value = numpy.random.randint(0,100)    
            
    
    # Register Python callback to UI changes
    # widget on_change event can only with bokeh server    
    my_slider_bokeh.on_change("value", UI_changed)    
    
    # Put widget python obj in bokeh doc
    doc.add_root(widgetbox(my_slider_bokeh))
    
    # Call periodic func to change widget python obj
    doc.add_periodic_callback(update_python_obj, 1000)


```

```python
# start bokeh server and app
app = show(modify_doc)
```
